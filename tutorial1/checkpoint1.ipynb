<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df["REVENUE_PER_STUDENT"] = df["TOTAL_REVENUE"] / df["GRADES_ALL_G"]
df["INSTRUCTION_EXPENDITURE_RATIO"] = df["INSTRUCTION_EXPENDITURE"] / df["TOTAL_EXPENDITURE"]
df["LOCAL_REVENUE_PER_STUDENT"] = df["LOCAL_REVENUE"] / df["GRADES_ALL_G"]
```

Feature engineering justification: 
- I created a revenue per student feature because similar to expenditure, revenue is something that differs a lot state by state, so normalizing this data point and seeing how much total funding affects students' test scores seemed to be key.
- I also made an instruction_expenditure_ratio feature because I thought it would be interesting to see whether the amount of the total expenditure that goes to instructional aid (like teachers) would affect test scores - if teachers are getting paid more/have more benefits, would they have better test scores than states that have a lower instruction expenditure ratio?
- I made a local revenue per student feature because states vary in their dependence on local revenue and I wanted to see if it would mean anything. Local revenue usually comes from property taxes (so places with higher property taxes might have higher local revenue), so it would be interesting to see how differing amounts of local revenue impact student test scores and it might create implications for wealthier localities.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["LOCAL_REVENUE_PER_STUDENT","REVENUE_PER_STUDENT","INSTRUCTION_EXPENDITURE_RATIO"]].dropna()
y = df.loc[X.index]["AVG_MATH_4_SCORE"]
y.fillna(y.median(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
      X, y, test_size= 0.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
# Out of the ones I looked at, Linear Regression worked the best
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
# mean error
print("Mean error: ", np.mean(model.predict(X_test)-y_test))

# mean absolute error
print("Mean Absolute Error: ", np.mean(np.abs(model.predict(X_test)-y_test)))

# root mean squared error
print("Root Mean Squared Error: ", np.mean((model.predict(X_test)-y_test)**2)**0.5)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = "LOCAL_REVENUE_PER_STUDENT"

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel("LOCAL_REVENUE_PER_STUDENT")
plt.title("Model Behavior On Training Set")
```

```python
col_name = "LOCAL_REVENUE_PER_STUDENT"

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel("LOCAL_REVENUE_PER_STUDENT")
plt.title("Model Behavior on Testing Set")
```
